### Data Visualization with Seaborn

- Seaborn is an extension of Matplotlib library that provides a high-level interface for drawing attractive and informative statistical graphics.

- Seaborn helps in complex visualization with just a few lines of code.

```python
# Importing Libraries : 
import seaborn as sns
```

```python
## Basic Plotting with Seaborn
# Seaborn has already pre-made dataset within itself we are using one of the dataset called 'tips'

tips = sns.load_dataset('tips')
tips
```

```python
## Creating a Scatterplot :
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex')
plt.title("Scatter Plot of Total Bill vs Tips")
```

```python
## Line Plot : 

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total Bill by Size")
plt.show()
```

```python
## Categorical Plot : 
# Bar Plot 

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
# Box Plot :
  
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
## Violin Plot : 

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
## Histograms : 

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False)
```

```python
## KDE Plot : 

sns.kdeplot(tips['total_bill'],fill=True)
plt.show()
```

```python
# Pairplot : 

sns.pairplot(tips)
```

```python
tips.dtypes
```

```python
## Heatmap :

corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd

sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
# Plot Total Sales by Product : 

plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Total Revenue',data=sales_df)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
